# imports 

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

# seaborn 특징 

`-` 특1: 입력으로 데이터프레임을 선호한다. (matplotlib은 array를 선호) 

- 그렇다고 해서 데이터프레임이 아닌 경우 그림이 아예 안 그려지지는 않는다. 
- 데이터프레임 형태는 long form 과 wide form 이 있다. [(ref)](https://seaborn.pydata.org/tutorial/data_structure.html#long-form-vs-wide-form-data) // 참고로 long form이 더 우수한 저장형태에요!
- wide-df = [array1,array2,array3]
- long-df = [array_val, array_cat]

`-` 특2: matplotlib을 존경함. ([ref](https://seaborn.pydata.org/index.html))

# [sns boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

`-` 데이터: 전북고등학교 

```python
y1=[75,75,76,76,77,77,79,79,79,98] # A선생님에게 통계학을 배운 학생의 점수들
y2=[76,76,77,77,78,78,80,80,80,81] # B선생님에게 통계학을 배운 학생의 점수들 
```

## plt 복습

```python
plt.boxplot([y1,y2]);
```

## sns wide df 

```python
df1=pd.DataFrame({1:y1,2:y2})
df1
```

`-` 예시1

```python
sns.boxplot(data=df1) 
#sns.boxplot(data=np.stack([y1,y2],axis=1))  # <- 잘 쓰진 않아요, 그냥 심심해서 해봤는데 되더라고요..?
```

## sns long df 

```python
df2=pd.DataFrame({'score':y1+y2,'class':['A']*len(y1)+['B']*len(y2)})
df2
```

`-` 예시1

```python
sns.boxplot(data=df2,x='class',y='score') 
```

## sns: array

`-` 예시1

```python
sns.boxplot(data=y1)
```

`-` 예시2

```python
sns.boxplot(y=y1)
```

`-` 예시3

```python
sns.boxplot(x=y1)
```

# [sns histplot](https://seaborn.pydata.org/generated/seaborn.histplot.html)

`-` 데이터

```python
x= np.random.randn(10000) 
y= np.random.randn(10000) +1
```

## plt 복습

`-` 예시1

```python
plt.hist(x,alpha=0.5)
plt.hist(y,alpha=0.5);
```

`-` 예시2

```python
plt.hist([x,y]);
```

## sns: wide df 

```python
df1=pd.DataFrame({'x':x,'y':y})
df1
```

`-` 예시1

```python
sns.histplot(data=df1)
```

`-` 예시2

```python
sns.histplot(data=df1,bins=20)
```

`-` 예시3

```python
sns.histplot(data=df1,bins=20,kde=True)
```

`-` 예시4

```python
sns.histplot(data=df1,bins=20,kde=True,element="step")
```

`-` 예시5

```python
sns.histplot(data=df1,bins=20,kde=True,element="step",lw=5) # mpl에 대한 존경심 확인 
```

## sns: long df 

```python
df2=pd.DataFrame({'val':np.concatenate([x,y]), 'var':['x']*len(x) + ['y']*len(y)})
df2
```

`-` 예시1

```python
sns.histplot(data=df2,x='val',hue='var',bins=20,kde=True,lw=0)
```

`-` 예시2

```python
sns.histplot(data=df2,y='val',hue='var',bins=20,lw=0,kde=True)
```

## sns: array

`-` 예시1

```python
sns.histplot(data=x)
```

`-` 예시2

```python
sns.histplot(x=x)
```

`-` 예시3

```python
sns.histplot(x=x,color='C0',bins=20,lw=0)
sns.histplot(x=y,color='C1',bins=20,lw=0)
```

# [sns lineplot](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

`-` data 

```python
np.random.seed(43052)
ϵ = np.random.randn(100)
```

```python
y = np.cumsum(ϵ)
```

## plt 복습

```python
plt.plot(ϵ,'--o')
```

```python
plt.plot(y,'--o')
```

## sns: array

`-` 예시1

```python
sns.lineplot(data=ϵ)
```

`-` 예시2

```python
sns.lineplot(data=y)
```

## sns: wide df

```python
df1=pd.DataFrame({'eps':ϵ, 'y':y})
df1
```

`-` 예시1

```python
sns.lineplot(data=df1)
```

`-` 예시2

```python
sns.lineplot(data=df1,dashes=False)
```

`-` 예시3

```python
# sns.lineplot(data=df1,dashes=[(3,1),(3,1)]) # 이코드는 최신버전의 sns에서 동작하지 않으므로 삭제합니다. (collab 에서는 정상동작)
```

`-` 예시4

```python
# sns.lineplot(data=df1,dashes=[(3,1),(15,3)],markers=['o','o'])  # 이코드는 최신버전의 sns에서 동작하지 않으므로 삭제합니다. (collab 에서는 정상동작)
```

## sns: long df

```python
df2= pd.DataFrame({'idx':list(range(100))*2,'val':np.concatenate([ϵ,y]),'cat':['eps']*100 + ['y']*100 })
df2
```

`-` 예시1

```python
sns.lineplot(data=df2, x='idx',y='val',hue='cat')
```

`-` 예시2

```python
sns.lineplot(data=df2, x='idx',y='val',style='cat',hue='cat',markers=True)
```

`-` 예시3

```python
sns.lineplot(data=df2, x='idx',y='val',style='cat',hue='cat',dashes=[(3,1),(3,3)],markers=['o','o'])
```
